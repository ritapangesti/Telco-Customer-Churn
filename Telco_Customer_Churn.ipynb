##1. Problem: Business Understanding, Data Understanding, and Objectives
Telco is a company that provided telecommunications services, such as home phone services and internet services (like Indih*me).

> Here are some of the services offered by Telco
- Home Phone Services, includes multiple lines.
- Internet Services that include No, DSL (Digital Subscriber Line), or Fiber-Optic connections.

Columns in Telco.csv table:
1. customerid: Unique ID for each customer
2. gender: To identify what gender the customer is, female or male
3. seniorcitizen: To find out whether customers are aged 65 years and over or not
4. partner: To know if the customer is married or not
5. dependents: To know if the customer lives with any dependents
6. tenure: To know the total amount of months that the customer has been with the company by the end of the quarter specified
7. phoneservice: To know if the customer subscribes to home phone service or not
8. multiplelines: To know if the customer subscribes to multiple phone lines
9. internetservice: To know if the customer subscribes to internet service or not
10. OnlineSecurity: To know if the customer subscribes to an additional online security service
11. OnlineBackup: To know if the customer subscribes to an additional online backup service
12. DeviceProtection: To know if the customer subscribes to an additional device protection or not
13. TechSupport: To know if the customer subscribes to an additional technical support plan with reduced wait times or not
14. streamingtv: To know if the customer uses their internet services to stream TV programs or not
15. streamingmovies: To know if the customer uses their internet services to stream movies or no
16. contract: To identify the customer's current contract type
17. Paperlessbilling: To know if the customer has chosen paperless billing
18. paymentmethod: To identify how the customer pays their bill
19. Monthlycharges: To know the customer's current total monthly charge for all their services from company
20. Totalcharges: To know the customer's total charges, calculated to the end of the quarter specified
21. Churn: To know the customer's loyalty, whether they left the company this quarter or not



##1. BUSINESS OBJECTIVES

1. Knowing the types of services that are highly demanded (Telco's TOP service)
2. Identifying the factors that cause customer churn:
- Knowing the influence of service types on customer churn among male and female customers
- Knowing the influence of contract types on customer churn among male and female customers
3. Giving recommendations or suggestions for Telco from the results

a. Dependents

```python
# Mencari modus Dependents
dependents_mode = df_telco['dependents'].mode()[0]
dependents_mode
```

##Conclussion: Insights and Recommendations

1. Highly Demanded Service Types: Phone Service and Internet Service, especially Fiber-Optic internet, are in high demand (Top/Populer) among customers.
2. Customer Churn Factors:


- Service Types: There is no significant difference in churn rates between phone service subscribers and non-phone service subscribers.
Internet service, particularly Fiber-Optic, needs to be evaluated in terms of quality, service, and other factors due to its high churn rate.
Customers who do not subscribe to additional services tend to experience higher churn rates.
- Contract Types: Month-to-month contracts may exhibit high non-churn rates, but their churn rates are relatively high compared to other contract types. This necessitates further evaluation to encourage subscribers to choose longer-term commitments.

3. Based on the insights from the analysis, Telco should consider to improving the quality of its internet service, especially Fiber Optic. Additionally, promoting additional services to customers who have not yet opted for them and offering longer contract options and attractive promotions can be beneficial. Apart from providing high-quality services, Telco is expected to create a sense of comfort for its customers. By doing so, customers are less likely to churn.
